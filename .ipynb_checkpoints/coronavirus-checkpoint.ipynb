```python
# Dependencies
import pandas as pd
import csv
from sqlalchemy import create_engine
```

```python
#importing datasets
recovered = pd.read_csv('Resources/time_series_19-covid-Recovered.csv')
deaths = pd.read_csv('Resources/time_series_19-covid-Deaths.csv')
confirmed = pd.read_csv('Resources/time_series_19-covid-Confirmed.csv')
```

```python
recovered.head()
recovered = recovered.drop(columns=['Lat', 'Long'])
df_recovered = pd.melt(recovered, id_vars=["Province/State", "Country/Region"])
df_recovered = df_recovered.rename(columns={"value": "Recovered Count", "variable": "Date"})
df_recovered.head()
```

```python
deaths.head()
deaths = deaths.drop(columns=['Lat', 'Long'])
df_deaths = pd.melt(deaths, id_vars=["Province/State", "Country/Region"])
df_deaths = df_deaths.rename(columns={"value": "Deaths Count", "variable": "Date"})
df_deaths.head()
```

```python
confirmed.head()
confirmed = confirmed.drop(columns=['Lat', 'Long'])
df_confirmed = pd.melt(confirmed, id_vars=["Province/State", "Country/Region"])
df_confirmed = df_confirmed.rename(columns={"value": "Confirmed Count", "variable": "Date"})
df_confirmed.head()
```

```python
df_combined = df_recovered.merge(df_deaths, on=("Province/State", "Country/Region", "Date"))
df_combined
```

```python
df_combined = df_combined.merge(df_confirmed, on=("Province/State", "Country/Region", "Date"))
df_combined
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/corona_db"
engine = create_engine(f'postgresql://{rds_connection_string}')


df_combined.to_sql(name='corona_virus', con=engine, if_exists='append', index=False)
```
