```python
import numpy as np 
import pandas as pd 
import plotly as py
import plotly.express as px
import csv
```

```python
#importing datasets
recovered = pd.read_csv('Resources/time_series_19-covid-Recovered.csv')
deaths = pd.read_csv('Resources/time_series_19-covid-Deaths.csv')
confirmed = pd.read_csv('Resources/time_series_19-covid-Confirmed.csv')
```

```python
recovered.head()
```

```python
fig = px.pie(recovered, values = 'Confirmed',names='Country/Region', height=600)
fig.update_traces(textposition='inside', textinfo='percent+label')

fig.update_layout(
    title_x = 0.5,
    geo=dict(
        showframe = False,
        showcoastlines = False,
    ))

fig.show()
```
